```python
# TASK 1
def is_palindrome(s: str) -> bool:
    cleaned_s = ''.join(char.lower() for char in s if char.isalpha())
    reversed_s = cleaned_s[::-1]
    return cleaned_s == reversed_s

print(is_palindrome(s='anade'))  
print(is_palindrome(s='ana'))     
```

```python
# TASK 2
import math

def solve_quad(a: float, b: float, c: float):
    if a == 0:
        if b == 0:
            return None
        else:
            root = -c / b
            return root
    else:
        discriminant = b**2 - 4*a*c
        
        if discriminant > 0:
            root1 = (-b + math.sqrt(discriminant)) / (2*a)
            root2 = (-b - math.sqrt(discriminant)) / (2*a)
            return root1, root2
        elif discriminant == 0:
            root = -b / (2*a)
            return root
        else:
            return None

print(solve_quad(a=1, b=-2, c=1))   
print(solve_quad(a=1, b=0, c=-1))   
print(solve_quad(a=0, b=2, c=-1))  
```

```python
# TASK 3
from typing import List

def merge(a: List[int], b: List[int]) -> List[int]:
    merged = [] 
    i = j = 0

    while i < len(a) and j < len(b):
        if a[i] <= b[j]:
            merged.append(a[i])
            i += 1
        else:
            merged.append(b[j])
            j += 1

    merged.extend(a[i:])
    merged.extend(b[j:])

    return merged

print(merge([1, 3, 3, 5], [3, 5, 5, 7])) 
```

```python
# TASK 4
from typing import List

def merge_sort(arr: List[int]) -> List[int]:
    if len(arr) <= 1:
        return arr

    middle = len(arr) // 2
    left_half = merge_sort(arr[:middle])
    right_half = merge_sort(arr[middle:])

    return merge(left_half, right_half)

def merge(left: List[int], right: List[int]) -> List[int]:
    merged = []
    left_index = right_index = 0

    while left_index < len(left) and right_index < len(right):
        if left[left_index] < right[right_index]:
            merged.append(left[left_index])
            left_index += 1
        else:
            merged.append(right[right_index])
            right_index += 1

    merged.extend(left[left_index:])
    merged.extend(right[right_index:])

    return merged

print(merge_sort([3, 1, 3, 5]))
```

```python
# TASK 9
import numpy as np

def sec_diag(X):
    rows, cols = X.shape
    # масив
    sec_diagonal = np.zeros(rows-1, dtype=X.dtype)
    for i in range(rows-1):
        sec_diagonal[i] = X[i, cols-i-2]
    return sec_diagonal

matrix = np.array([[1, 2], [3, 4]])
result = sec_diag(matrix)
print(result) 
```

```python
# TASK 11
import pandas as pd

# хавантаження датасету
dataset = pd.read_csv('titanic.csv')

# середній вік пасажирів
mean_age_all = dataset['Age'].mean()
print("Середній вік пасажирів:", mean_age_all)

# середній вік чоловіків
mean_age_male = dataset[dataset['Sex'] == 'male']['Age'].mean()
print("Середній вік чоловіків:", mean_age_male)

# середній вік жінок
mean_age_female = dataset[dataset['Sex'] == 'female']['Age'].mean()
print("Середній вік жінок:", mean_age_female)

# максимальний вік серед чоловіків і жінок
max_age_male = dataset[dataset['Sex'] == 'male']['Age'].max()
max_age_female = dataset[dataset['Sex'] == 'female']['Age'].max()
print("Максимальний вік серед чоловіків:", max_age_male)
print("Максимальний вік серед жінок:", max_age_female)

# мінімальнний вік серед чоловіків і жінок
min_age_male = dataset[dataset['Sex'] == 'male']['Age'].min()
min_age_female = dataset[dataset['Sex'] == 'female']['Age'].min()
print("Мінімальний вік серед чоловіків:", min_age_male)
print("Мінімальний вік серед жінок:", min_age_female)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f1(x):
    return np.sin(x) / x

def f2(x):
    return np.cos(x) / (x**2 + 1)

def f3(x):
    return np.cos(x)**2 - np.sin(x) / 2

def f4(x):
    return np.cos(x)

def f5(x):
    return np.sin(x) / (x**2 + 1)

x = np.linspace(-10, 10, 400)

plt.figure(figsize=(10, 8))

plt.plot(x, f1(x), label=r'$\frac{sin(x)}{x}$')
plt.plot(x, f2(x), label=r'$\frac{cos(x)}{x^2 + 1}$')
plt.plot(x, f3(x), label=r'$cos^2(x) - \frac{sin(x)}{2}$')
plt.plot(x, f4(x), label=r'$cos(x)$')
plt.plot(x, f5(x), label=r'$\frac{sin(x)}{x^2 + 1}$')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Графіки функцій')
plt.legend()

plt.grid(True)
plt.show()
```
